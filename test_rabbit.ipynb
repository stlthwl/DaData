```python
# функция для получения данных из таблицы "offices_source" базы данных источника
def get_data(arg1, arg2, arg3, arg4, arg5):
    con = psycopg2.connect(
        database=arg1,
        user=arg2,
        password=arg3,
        host=arg4,
        port=arg5
    )
    print("Database opened successfully")
    cur = con.cursor()
    cur.execute("SELECT * from offices_source")
    data = cur.fetchall()
    print("Operation done successfully")
    con.commit()
    con.close()
    return data
```

```python
# загрузка данных в dataframe
res = get_data(db_name, db_user, db_pass, db_host, db_port)
df = pd.DataFrame(res)
df
```

```python
# ограничение выборки 10-ю записями для теста
test_df = df.head(10)
test_df
```

```python
# функция для получения широты и долготы по API DaData
def get_location(address):
    token = api_key
    secret = secret_key
    dadata = Dadata(token, secret)
    result = dadata.clean("address", address)
    return result
```

```python
# Сбор сведений о широте и долготе по адресам

lon_lst = []
lat_lst = []

for address in list(test_df[4]):
    address_info = get_location(address)
    lon_lst.append(address_info['geo_lon'])
    lat_lst.append(address_info['geo_lat'])  
```

```python
# проверка длины списков

if  (len(lon_lst) + len(lat_lst)) / 2 == len(list(test_df[4])):
    print(f"Соответствие длинны списков с длиной таблицы с адресами \n{len(lon_lst), len(lat_lst)}")
```

```python
# результирующий набор данных, включая долготу и широту
res_df = pd.DataFrame()

res_df['#'] = test_df[0]
res_df['uuid'] = test_df[3]
res_df['address'] = test_df[4]
res_df['partner_title'] = test_df[6]
res_df['office_title'] = test_df[7]
res_df['phone'] = test_df[8]
res_df['market_segment'] = test_df[9]
res_df['longitude'] = lon_lst
res_df['latitude'] = lat_lst

res_df
```
